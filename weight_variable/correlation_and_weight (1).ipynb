```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
plt.rc('font', family='NanumGothic')
matplotlib.rcParams['axes.unicode_minus'] = False

file = pd.read_csv('C:\\신현호\\2021_2학기_공모전\\코드구성\\상관관계분석\\구별_통합_데이터_02.csv',
                  encoding='utf-8')
file
```

```python
ppl = file['인구수평균']
charger = file['충전소 수']
car = file['전기차 수']
sol = file['태양광 발전소 수']
elec = file['전기사용량 평균']
ref_car = file['전기차 변수']
wind = file['풍속 비율']
df = pd.DataFrame()
df['인구수평균'] = ppl
df['충전소 수'] = charger
df['전기차 수'] = car
df['전기차 변수'] = ref_car
df['태양광 발전소 수'] = sol
df['전기사용량 평균'] = elec
df['풍력 비율'] = wind
df
```

```python
# df['인구수평균'] = ppl
# df['충전소 수'] = charger
# df['전기차 수'] = car
# df['태양광 발전소 수'] = sol
# df['전기사용량 평균'] = elec

from sklearn.preprocessing import MinMaxScaler

min_max_scaler = MinMaxScaler(feature_range=(0,1))

ppl_scaled = min_max_scaler.fit_transform(df['인구수평균'].values.reshape(-1,1))
#charger_scaled = min_max_scaler.fit_transform(df['충전소 수'].values.reshape(-1,1))
#car_scaled = min_max_scaler.fit_transform(df['전기차 수'].values.reshape(-1,1))
ref_scaled = min_max_scaler.fit_transform(df['전기차 변수'].values.reshape(-1,1))
sol_scaled = min_max_scaler.fit_transform(df['태양광 발전소 수'].values.reshape(-1,1))
#elec_scaled = min_max_scaler.fit_transform(df['전기사용량 평균'].values.reshape(-1,1))
wind_scaled = min_max_scaler.fit_transform(df['풍력 비율'].values.reshape(-1,1))

scaled_df = pd.DataFrame()

scaled_df['인구수평균'] = np.array(ppl_scaled).flatten().tolist()
#scaled_df['충전소 수'] = np.array(charger_scaled).flatten().tolist()
#scaled_df['전기차 수'] = np.array(car_scaled).flatten().tolist()
scaled_df['전기차 변수'] = np.array(ref_scaled).flatten().tolist()
scaled_df['태양광 발전소 수'] = np.array(sol_scaled).flatten().tolist()
#scaled_df['전기사용량 평균'] = np.array(elec_scaled).flatten().tolist()
scaled_df['풍력 비율'] = np.array(wind_scaled).flatten().tolist()
scaled_df
```

```python
s_corr_df = scaled_df.corr()
s_corr_df = s_corr_df.apply(lambda x: round(x,2))
s_corr_df
```

```python
fig, ax = plt.subplots()
im = ax.imshow(s_corr_df, cmap='GnBu')

cbar = ax.figure.colorbar(im, ax=ax)

ax.set_xticks(np.arange(len(s_corr_df.columns)))
ax.set_yticks(np.arange(len(s_corr_df.index)))

ax.set_xticklabels(s_corr_df.columns,rotation='40')
ax.set_yticklabels(s_corr_df.columns)

for x in range(len(s_corr_df.columns)):
    for y in range(len(s_corr_df.index)):
        ax.text(y,x,s_corr_df.iloc[y,x],ha='center',va='center',color='black')
       
fig.tight_layout()
plt.show()
```

```python
# scaled_df['인구수평균'] = np.array(ppl_scaled).flatten().tolist()
# scaled_df['충전소 수'] = np.array(charger_scaled).flatten().tolist()
# scaled_df['전기차 수'] = np.array(car_scaled).flatten().tolist()
# #scaled_df['전기차 변수'] = np.array(ref_scaled).flatten().tolist()
# scaled_df['태양광 발전소 수'] = np.array(sol_scaled).flatten().tolist()
# scaled_df['전기사용량 평균'] = np.array(elec_scaled).flatten().tolist()
import math

k = 1/(math.log(25)) #k=1/(ln(m)
ppl_scaled_list = np.array(ppl_scaled).flatten().tolist()
#entropy_ppl = []
ppl_ent = 0

for i in ppl_scaled_list:
    try:
        ppl_ent = ppl_ent+(i*math.log(i))
    except ValueError:
        ppl_ent = ppl_ent+0

ppl_ent = (-k)*ppl_ent
ppl_div = 1-ppl_ent
ppl_div
```

```python
sol_scaled_list = np.array(sol_scaled).flatten().tolist()
sol_ent = 0

for i in sol_scaled_list:
    try:
        sol_ent = sol_ent+(i*math.log(i))
    except ValueError:
        sol_ent = sol_ent+0

sol_ent = (-k)*sol_ent
sol_div = 1-sol_ent
sol_div
```

```python
ref_scaled_list = np.array(ref_scaled).flatten().tolist()
ref_ent = 0

for i in ref_scaled_list:
    try:
        ref_ent = ref_ent+(i*math.log(i))
    except ValueError:
        ref_ent = ref_ent+0

ref_ent = (-k)*ref_ent
ref_div = 1-ref_ent
ref_div
```

```python
wind_scaled_list = np.array(wind_scaled).flatten().tolist()
wind_ent = 0

for i in wind_scaled_list:
    try:
        wind_ent = wind_ent+(i*math.log(i))
    except ValueError:
        wind_ent = wind_ent+0

wind_ent = (-k)*wind_ent
wind_div = 1-wind_ent
wind_div
```

```python
#전기차 변수 가중치
ref_w = ref_div/(sol_div+ref_div+ppl_div+wind_div)
```

```python
#인구 수 가중치
ppl_w  = ppl_div/(sol_div+ref_div+ppl_div+wind_div)
```

```python
#태양광 수 가중치
sol_w = sol_div/(sol_div+ref_div+ppl_div+wind_div)
```

```python
#풍속 가중치
wind_w = wind_div/(sol_div+ref_div+ppl_div+wind_div)
```

```python
print("전기 차 변수 가중치:{}".format(round(ref_w,2)))
print("인구 수 변수 가중치:{}".format(round(ppl_w,2)))
print("태양광 변수 가중치:{}".format(round(sol_w,2)))
print("풍속 변수 가중치:{}".format(round(wind_w,2)))
```

```python
result = []
for i in range(25):
    a = scaled_df.loc[i,'인구수평균']
    b = scaled_df.loc[i,'전기차 변수']
    c = scaled_df.loc[i,'태양광 발전소 수']
    d = scaled_df.loc[i,'풍력 비율']
    result.append(round(a*ppl_w+b*ref_w+c*sol_w+d*wind_w,3))
    
result
```
