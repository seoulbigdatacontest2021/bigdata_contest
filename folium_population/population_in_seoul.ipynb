```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import folium
import json

ppl = pd.read_csv('C:\\신현호\\2021_2학기_공모전\\코드구성\\인구수_folium\\population_avg.csv',encoding='utf-8')
ppl
```

```python
#for coordinates for folium
code_elec_car__ = pd.read_csv('C:\\신현호\\2021_2학기_공모전\codecode.csv',encoding='cp949')
```

```python
for i in range(len(ppl)):
    ppl.loc[i,"코드"] = str(ppl.loc[i,"코드"])
```

```python
lat = code_elec_car__['y'].mean()
long = code_elec_car__['x'].mean()

m = folium.Map([lat,long],zoom_start=10,tiles="cartodbpositron")

def folium_del_legend(choropleth: folium.Choropleth): 
    del_list = [] 
    for child in choropleth._children: 
        if child.startswith('color_map'): 
            del_list.append(child) 
    for del_item in del_list: 
        choropleth._children.pop(del_item) 
    return choropleth
```

```python
geo = json.load(open('C:\\Users\\hhshi\\OneDrive\\바탕 화면\\seoul_municipalities_geo_simple.json',
                    encoding='utf-8'))
```

```python
k = folium.Choropleth(geo_data = geo,
             data=ppl,
             columns=['코드','인구수평균'],
             fill_color= 'Oranges',
             key_on='properties.code',
             legend_name='각 구별 인구 수').add_to(m)
folium_del_legend(k)
m.save('seoul_population.html')
m
```
